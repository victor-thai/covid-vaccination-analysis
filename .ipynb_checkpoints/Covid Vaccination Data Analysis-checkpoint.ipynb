```python
# look at which vaccine is most popular
vaccinations.groupby('vaccine')['total_vaccinations'].sum().sort_values(ascending = False)
```

As we look into which vaccine is most popular among all people that are vaccinated, we can see that Pfizer, Moderna, and AstraZeneca are the most widely used among all countries.

```python
# look into which vaccine is most popular among each country
country_vaccine = vaccinations.groupby(['location','vaccine']).sum().sort_values(['location', 'total_vaccinations'], ascending = [True, False]).reset_index()
country_vaccine_popular = country_vaccine.groupby('location').first()
country_vaccine_popular
```

```python
# most popular vaccine among other popular vaccines in other countries
country_vaccine_popular.value_counts('vaccine')
```

Through our analysis, Pfizer seems to be the vaccine that dominates over all the other popular vaccines in other countries.
